# Bar chart,Area chart ,Histogram for canada immigration dataset

```python
import numpy as np
import pandas as pd
!pip3 install openpyxl
```

```python
df_canada=pd.read_excel('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
                        sheet_name='Canada by Citizenship',
                        skiprows=range(20),
                        skipfooter=2)
print('data downloaded and read into data frame')
```

```python
df_canada.head()
```

```python
df_canada.drop(columns={'Type','AREA','REG','DEV','Coverage'},axis=1,inplace=True)
df_canada.head()
```

```python
df_canada.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
df_canada.head()
```

```python
all(isinstance(column,str)for column in df_canada.columns)
```

```python
df_canada.columns=list(map(str,df_canada.columns))
all(isinstance(column,str)for column in df_canada.columns)
```

```python
df_canada.set_index('Country',inplace=True)
df_canada['Total']=df_canada.sum(axis=1)
df_canada.head(3)
```

```python
years=list(map(str,range(1980,2013)))
years
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Area plot/Stacked line plot

```python
df_canada.sort_values(['Total'],ascending=False,axis=0,inplace=True)
df_top5=df_canada.head()
df_top5
```

```python
df_top5.plot(kind='area')
plt.title('Top 5 immigrants to canada')
plt.ylabel('Total of immigrants')
plt.xlabel('country')
plt.show()
```

```python
df_top5=df_top5[years].transpose()
df_top5.plot(kind='area', stacked=False)
plt.title('Top 5 immigrants to canada')
plt.ylabel('Total of immigrants')
plt.xlabel('years')
plt.show()
```

when we keep stacked=false it will become unstacked plot

```python
# unstacked plot has default alpha
df_top5.plot(kind='area', 
             alpha=0.25,  # 0 - 1, default value alpha = 0.5
             stacked=False,
             figsize=(20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

artist layer/Object oreintentation method

```python
ax=df_top5.plot(kind='area', 
             alpha=0.25,  # 0 - 1, default value alpha = 0.5
             stacked=False,
             figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

## Histograms
What is the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?

```python
df_canada['2013'].head()
```

```python
count, bin_edges = np.histogram(df_canada['2013'])

print(count)
print(bin_edges)
```

```python
df_canada['2013'].plot(kind='hist', figsize=(8, 5))

# add a title to the histogram
plt.title('Histogram of Immigration from 195 Countries in 2013')
# add y-label
plt.ylabel('Number of Countries')
# add x-label
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_canada['2013'])

df_canada['2013'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Immigration from 195 countries in 2013') # add a title to the histogram
plt.ylabel('Number of Countries') # add y-label
plt.xlabel('Number of Immigrants') # add x-label

plt.show()
```

What is the immigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013?

```python
df_canada.loc[['Denmark', 'Norway', 'Sweden'], years]
df_canada.loc[['Denmark', 'Norway', 'Sweden'], years].plot.hist()
```

```python
df_t=df_canada.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist')
plt.title(' Histogram of immigration population from denmark,norway,sweden')
plt.ylabel('population')
plt.xlabel('years')
plt.show()
```

```python
count,bin_edges=np.histogram(df_t,15)
df_t.plot(kind ='hist', 
          figsize=(10, 6),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title(' Histogram of immigration population from denmark,norway,sweden')
plt.xlabel('population')
plt.ylabel('years')
plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10   #  first bin value is 31.0, adding buffer of 10 for aesthetic purposes 
xmax = bin_edges[-1] + 10  #  last bin value is 308.0, adding buffer of 10 for aesthetic purposes

# stacked Histogram
df_t.plot(kind='hist',
          figsize=(10, 6), 
          bins=15,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

Use the scripting layer to display the immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013? Use an overlapping plot with 15 bins and a transparency value of 0.35.

```python
df_t1=df_canada.loc[['Greece', 'Albania', 'Bulgaria'], years].transpose()
df_t1.head()
count, bin_edges = np.histogram(df_t1, 15)
xmin = bin_edges[0] - 10   
xmax = bin_edges[-1] + 10 

# stacked Histogram
df_t1.plot(kind='hist',
          figsize=(10, 6), 
          bins=15,
          xticks=bin_edges,
           alpha=0.35,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
         )

plt.title('Histogram of Immigration from Greece,Albania,Bulgaria from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

## Bar Charts(Dataframe)
kind=bar-vertical
kind=barh-horizontal

 compare the number of Icelandic immigrants (country = 'Iceland') to Canada from year 1980 to 2013.

```python
df_i=df_canada.loc['Iceland',years]
df_i.plot(kind='bar')
plt.title('number of immigrants from iceland')
plt.ylabel('immigrants')
plt.xlabel('years')
plt.show()
```

```python
df_i.plot(kind='bar', figsize=(10, 6), rot=90)  # rotate the xticks(labelled points on x-axis) by 90 degrees

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('',  # s: str. Will leave it blank for no text
             xy=(32, 70),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

plt.show()
```

```python
df_i.plot(kind='bar', figsize=(10, 6), rot=90)

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('',  # s: str. will leave it blank for no text
             xy=(32, 70),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis',  # text to display
             xy=(28, 30),  # start the text at at point (year 2008 , pop 30)
             rotation=72.5,  # based on trial and error to match the arrow
             va='bottom',  # want the text to be vertically 'bottom' aligned
             ha='left',  # want the text to be horizontally 'left' algned.
             )

plt.show()
```

Using the scripting later and the `df_canada` dataset, create a *horizontal* bar plot showing the *total* number of immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total immigrant count

```python
df_canada.sort_values(['Total'],ascending=False,axis=0,inplace=True)
df_top15=df_canada.head(15)
df_top15
```

```python
df_top15.plot(kind='barh', figsize=(12, 12), color='steelblue')

plt.xlabel('Number of Immigrants')
plt.title('top 15 Immigrants to Canada from 1980 to 2013')

for index, value in enumerate(df_top15): 
        label = format(int(value), ',') # format int with commas
    
    # place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
        plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')

    
plt.show()
```
